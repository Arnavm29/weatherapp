```python
#Libraries#
from tkinter import *
import time
import requests as req
import pyttsx3

global widgets
widgets = []

#Functions#
#Speak Text
def speakText(command):
    ttsbot = pyttsx3.init()
    ttsbot.say(command)
    ttsbot.runAndWait()
    
#Delete Widgets
def delWidgets():
    global widgets
    for widget in widgets:
        widget.destroy()
    widgets = []
    
#Weather using API
def getCurrentWeather():
    city = cityEntry.get()
    api = "https://api.openweathermap.org/data/2.5/weather?q=" + \
      city + "&appid=e80d64f455b6591216879f79c4621514"
    json_data = req.get(api).json()
    description = str(json_data["weather"][0]["description"])
    temp = int(json_data['main']['temp'] - 273.15)
    pressure = str(json_data['main']['pressure'])
    humidity = str(json_data['main']['humidity'])
    sunrise = str(time.strftime('%I:%M:%S', time.gmtime(json_data['sys']['sunrise'] - 21600)))
    sunset = str(time.strftime('%I:%M:%S', time.gmtime(json_data['sys']['sunset'] - 21600)))
    w_speed = str(json_data["wind"]["speed"])
    res = description + "\n" + str(temp) + "°C" + "\n" + "Pressure: " + str(pressure) + " milibars" + "\n" + "Humidity: " + str(humidity) + "%" + "\n" + "Sunrise: " + sunrise + " a.m." + "\n" + "Sunset: " + sunset + " p.m." +"\n" + "Wind Speed: " + str(w_speed) + " miles per hour"
    weatherLabel.config(text=res)

def getCityGeoCode(city):
    api = "http://api.openweathermap.org/geo/1.0/direct?q=" + \
      city + "&appid=e80d64f455b6591216879f79c4621514"
    json_data = req.get(api).json()
    lat = str(json_data[0]['lat'])
    lon = str(json_data[0]['lon'])
    return [lat,lon]

def getWeatherForecast():
    city = cityEntry.get()
    coord = getCityGeoCode(city)
    api = "https://api.openweathermap.org/data/2.5/forecast?lat=" + coord[0] + "&lon=" + coord[1] + "&appid=e80d64f455b6591216879f79c4621514"
    json_data = req.get(api).json()
    description = str(json_data['list'][0]["weather"][0]["description"])
    temp = int(json_data['list'][0]['main']['temp'] - 273.15)
    pressure = str(json_data['list'][0]['main']['pressure'])
    humidity = str(json_data['list'][0]['main']['humidity'])
    w_speed = str(json_data['list'][0]["wind"]["speed"])
    res = description + "\n" + str(temp) + "°C" + "\n" + "Pressure: " + str(pressure) + " milibars" + "\n" + "Humidity: " + str(humidity) + "%" + "\n" +  "Wind Speed: " + str(w_speed) + " miles per hour"
    weatherLabel.config(text=res)

#Main Menu
def mainMenu():
    root.title("Main Menu")
    
    #Delete widgets prior
    delWidgets()

    #Title Label
    titleLabel = Label(root, text="Weather App", bg="#ADD8E6", fg="#000000", font=style2)
    titleLabel.grid(row=0, column=0, columnspan=2, pady=10)
    widgets.append(titleLabel)
    
    #Buttons
    currentWeatherButton = Button(root, text="Current Weather", command=getCurrentWeatherWindow, bg="#FFA07A", font=style1, relief=RAISED, borderwidth=0, highlightthickness=0, padx=10, pady=5)
    currentWeatherButton.grid(row=1, column=0, padx=20, pady=10)
    widgets.append(currentWeatherButton)

    weatherForecastButton = Button(root, text="Weather Forecast", command=getWeatherForecastWindow, bg="#FFA07A", font=style1, relief=RAISED, borderwidth=0, highlightthickness=0, padx=10, pady=5)
    weatherForecastButton.grid(row=1, column=1, padx=20)
    widgets.append(weatherForecastButton)
    
#CurrentWeatherWindow
def getCurrentWeatherWindow():

    root.title("Current Weather")
    
    #Delete widgets prior
    delWidgets()

    #Title Label
    titleLabel = Label(root, text="Current Weather", bg="#ADD8E6", fg="#000000", font=style2)
    titleLabel.grid(row=0, column=0, columnspan=2, pady=10)
    widgets.append(titleLabel)
    
    #City Label and Entry
    cityLabel = Label(root, text="Enter City:", bg="#ADD8E6", fg="#000000", font=style1)
    cityLabel.grid(row=1, column=0, padx=10, pady=10)
    widgets.append(cityLabel)

    global cityEntry
    cityEntry = Entry(root, font=style1)
    cityEntry.grid(row=1, column=1, padx=10, pady=10)
    widgets.append(cityEntry)

    #Submit Button
    submitButton = Button(root, text="Enter", command=getCurrentWeather, bg="#FFA07A", font=style1, relief=RAISED, borderwidth=0, highlightthickness=0, padx=10, pady=5)
    submitButton.grid(row=2, column=0, columnspan=2)
    widgets.append(submitButton)

    #Back Button
    submitButton = Button(root, text="Back", command=mainMenu, bg="#FFA07A", font=style1, relief=RAISED, borderwidth=0, highlightthickness=0, padx=10, pady=5)
    submitButton.grid(row=3, column=2, columnspan=2)
    widgets.append(submitButton)
    
    #Weather Label
    global weatherLabel
    weatherLabel = Label(root, text="", bg="#ADD8E6", fg="#000000", font=style1)
    weatherLabel.grid(row=3, column=0, columnspan=2, padx=10, pady=10)
    widgets.append(weatherLabel)

    root.after(100, speakText, "Current Weather")

#CurrentWeatherWindow
def getWeatherForecastWindow():

    root.title("Forecast Weather")
    
    #Delete widgets prior
    delWidgets()

    #Title Label
    titleLabel = Label(root, text="Weather Forecast", bg="#ADD8E6", fg="#000000", font=style2)
    titleLabel.grid(row=0, column=0, columnspan=2, pady=10)
    widgets.append(titleLabel)

    #City Label and Entry
    cityLabel = Label(root, text="Enter City:", bg="#ADD8E6", fg="#000000", font=style1)
    cityLabel.grid(row=1, column=0, padx=10, pady=10)
    widgets.append(cityLabel)

    global cityEntry
    cityEntry = Entry(root, font=style1)
    cityEntry.grid(row=1, column=1, padx=10, pady=10)
    widgets.append(cityEntry)

    #Submit Button
    submitButton = Button(root, text="Enter", command=getWeatherForecast, bg="#FFA07A", font=style1, relief=RAISED, borderwidth=0, highlightthickness=0, padx=10, pady=5)
    submitButton.grid(row=2, column=0, columnspan=2)
    widgets.append(submitButton)

    #Back Button
    submitButton = Button(root, text="Back", command=mainMenu, bg="#FFA07A", font=style1, relief=RAISED, borderwidth=0, highlightthickness=0, padx=10, pady=5)
    submitButton.grid(row=3, column=2, columnspan=2)
    widgets.append(submitButton)
    
    #Weather Label
    global weatherLabel
    weatherLabel = Label(root, text="", bg="#ADD8E6", fg="#000000", font=style1)
    weatherLabel.grid(row=3, column=0, columnspan=2, padx=10, pady=10)
    widgets.append(weatherLabel)

    root.after(100, speakText, "Weather Forecast")
    
#Tkinter#
#Root Window#
root = Tk()
root.title("Main Menu")
root.geometry('400x300')
root.configure(bg="#ADD8E6")
root.resizable(False,False)

#Style
style1 = ("Helvetica", 12)
style2 = ("Helvetica", 20, "bold")

mainMenu()
root.after(100, speakText, "welcome to my weather app")
root.mainloop()
```

```python
from tkinter import *
root = Tk()
root.title("Current Weather")
root.geometry('300x300')
root.configure(bg="#ADD8E6")
root.resizable(False,False)


       
def openCurrentWeatherWindow():
    #Tkinter#
    #Root Window
    root = Tk()
    root.title("Current Weather")
    root.geometry('300x300')
    root.configure(bg="#ADD8E6")
    root.resizable(False,False)

    #Style
    style1 = ("Helvetica" , 12)

    #Title Label
    titleLabel = Label(root, text="Weather App", bg="#ADD8E6", fg="#000000", font=("Helvetica", 20, "bold"))
    titleLabel.grid(row=0, column=0, columnspan=2, pady=10)

    #City Label and Entry
    cityLabel = Label(root, text="Enter City:", bg="#ADD8E6", fg="#000000", font=style1)
    cityLabel.grid(row=1, column=0, padx=10, pady=10)

    cityEntry = Entry(root, font=style1)
    cityEntry.grid(row=1, column=1, padx=10, pady=10)

    #Submit Button
    submitButton = Button(root, text="Enter", command=getCurrentWeather, bg="#FFA07A", font=style1, relief=RAISED, borderwidth=0, highlightthickness=0, padx=10, pady=5)
    submitButton.grid(row=2, column=0, columnspan=2)

    #Weather Label
    weatherLabel = Label(root, text="", bg="#ADD8E6", fg="#000000", font=style1)
    weatherLabel.grid(row=3, column=0, columnspan=2, padx=10, pady=10)
    
titleLabel = Label(root, text="Weather App", bg="#ADD8E6", fg="#000000", font=("Helvetica", 20, "bold"))
titleLabel.grid(row=0, column=0, pady=10, padx=20)

#Buttons
currentWeatherButton = Button(root, text="Current Weather", command=openCurrentWeatherWindow, bg="#FFA07A", font=style1, relief=RAISED, borderwidth=0, highlightthickness=0, padx=10, pady=5)
currentWeatherButton.grid(row=1, column=0, padx=20, pady=10)

weatherForecastButton = Button(root, text="Weather Forecast", command=openCurrentWeatherWindow, bg="#FFA07A", font=style1, relief=RAISED, borderwidth=0, highlightthickness=0, padx=10, pady=5)
weatherForecastButton.grid(row=2, column=0, padx=20)

root.mainloop()
```
